## Running the OCR Model

```python
import easyocr
import re
import os
import pandas as pd

reader = easyocr.Reader(['en'])

def extract_code_from_image(image_path):
    results = reader.readtext(image_path, detail=0)
    full_text = ' '.join(results)
    codes = re.findall(r'\b\d{6,8}\b', full_text)
    return codes, full_text

# Step 1: Build image → code mapping
image_code_map = []  # list of dicts

folders = ['1', '2', '3', '4']
for folder in folders:
    for file in os.listdir(folder):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(folder, file)
            try:
                codes, full_text = extract_code_from_image(image_path)
                image_code_map.append({
                    'image_path': image_path,
                    'detected_codes': codes,
                    'first_code': codes[0] if codes else None,  # use first found code
                    'full_text': full_text
                })
            except Exception as e:
                print(f"Error in {file}: {e}")

# Convert to DataFrame
mapping_df = pd.DataFrame(image_code_map)
print(mapping_df[['image_path', 'first_code']].head(20))
```

```python
# Step 2: Load your sales CSV
sales_df = pd.read_csv('AI ML Task Sheet - sales data.csv')
print(sales_df.head())

# Make sure code column is string type in both DataFrames
mapping_df['first_code'] = mapping_df['first_code'].astype(str)
sales_df['code'] = sales_df['code'].astype(str)

# Aggregate sales per code (sum qty, mean rate)
sales_agg = sales_df.groupby('code').agg(
    total_qty=('qty', 'sum'),
    avg_rate=('rate', 'mean'),
    num_orders=('qty', 'count')
).reset_index()

# Step 3: Merge image mapping with sales data
final_df = mapping_df.merge(sales_agg, left_on='first_code', right_on='code', how='left')

print(f"\nTotal images: {len(final_df)}")
print(f"Successfully matched: {final_df['total_qty'].notna().sum()}")
print(f"Unmatched (no sales data): {final_df['total_qty'].isna().sum()}")

print(final_df[['image_path', 'first_code', 'total_qty', 'avg_rate']].head(20))
```

```python
# Save full mapping to CSV for inspection
final_df.to_csv('image_sales_mapping.csv', index=False)
print("\nSaved to image_sales_mapping.csv")

# See unmatched ones - to debug OCR issues
unmatched = final_df[final_df['total_qty'].isna()]
print("\nUnmatched images:")
print(unmatched[['image_path', 'first_code', 'full_text']])
```

```python
# Keep only matched rows (where OCR code was found in CSV)
matched_df = final_df[final_df['total_qty'].notna()].copy()

# Drop rows where first_code is None (OCR failed completely)
matched_df = matched_df[matched_df['first_code'].notna()].copy()

# Reset index
matched_df = matched_df.reset_index(drop=True)

# Clean up - keep only useful columns
matched_df = matched_df[['image_path', 'first_code', 'total_qty', 'avg_rate', 'num_orders']]

# Save it
matched_df.to_csv('matched_data.csv', index=False)

print(f"Clean matched DataFrame: {len(matched_df)} rows")
print(matched_df.head())
```

## Extracting Image Embeddings

```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import numpy as np
import pandas as pd

# Load pretrained MobileNetV2 (remove final classification layer)
model = models.mobilenet_v2(pretrained=True)
model.classifier = torch.nn.Identity()  # output: 1280-dim embedding
model.eval()

# Image preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

def get_image_embedding(image_path):
    try:
        img = Image.open(image_path).convert('RGB')
        tensor = transform(img).unsqueeze(0)  # add batch dim
        with torch.no_grad():
            embedding = model(tensor).squeeze().numpy()  # shape: (1280,)
        return embedding
    except Exception as e:
        print(f"Error with {image_path}: {e}")
        return None

# Load your matched data
matched_df = pd.read_csv('matched_data.csv')

# Extract embeddings for all 102 images
print("Extracting embeddings... (takes 1-2 mins)")
embeddings = []
valid_indices = []

for idx, row in matched_df.iterrows():
    emb = get_image_embedding(row['image_path'])
    if emb is not None:
        embeddings.append(emb)
        valid_indices.append(idx)

# Keep only rows where embedding succeeded
matched_df = matched_df.loc[valid_indices].reset_index(drop=True)
embeddings = np.array(embeddings)

print(f"Embeddings shape: {embeddings.shape}")  # should be (102, 1280)
print(f"Valid rows: {len(matched_df)}")
```

```python
from sklearn.preprocessing import StandardScaler

# Target variable
y = matched_df['total_qty'].values

# Optional extra features from matched_df
extra_features = matched_df[['avg_rate']].fillna(0).values

# Combine image embeddings + extra features
X = np.hstack([embeddings, extra_features])  # shape: (102, 1282)

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"y range: {y.min()} to {y.max()}")
```

## Training Model

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import xgboost as xgb
import numpy as np

# Log transform the target (handles skew and outliers)
y_log = np.log1p(matched_df['total_qty'].values)

# Try different PCA components
pca = PCA(n_components=30)  # reduced from 50
X_reduced = pca.fit_transform(X)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_reduced)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_log, test_size=0.2, random_state=42
)

# Try 3 models
models_to_try = {
    'RandomForest': RandomForestRegressor(
        n_estimators=200,
        max_depth=5,
        min_samples_leaf=4,
        max_features='sqrt',
        random_state=42
    ),
    'GradientBoosting': GradientBoostingRegressor(
        n_estimators=100,
        max_depth=3,
        learning_rate=0.05,
        min_samples_leaf=4,
        random_state=42
    ),
    'XGBoost': xgb.XGBRegressor(
        n_estimators=100,
        max_depth=3,
        learning_rate=0.05,
        subsample=0.8,
        random_state=42,
        verbosity=0
    )
}

print("=== Model Comparison ===\n")
best_model = None
best_mae = float('inf')

for name, m in models_to_try.items():
    m.fit(X_train, y_train)
    y_pred_log = m.predict(X_test)

    # Convert back from log scale
    y_pred_actual = np.expm1(y_pred_log)
    y_test_actual = np.expm1(y_test)

    mae = mean_absolute_error(y_test_actual, y_pred_actual)
    r2  = r2_score(y_test_actual, y_pred_actual)

    # Cross-val score on full data (more reliable with small dataset)
    cv_scores = cross_val_score(m, X_scaled, y_log, cv=5, scoring='r2')

    print(f"[{name}]")
    print(f"  MAE          : {mae:.2f} units")
    print(f"  R²           : {r2:.4f}")
    print(f"  CV R² (5fold): {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
    print()

    if mae < best_mae:
        best_mae = mae
        best_model = (name, m)

print(f"Best model: {best_model[0]} with MAE: {best_mae:.2f}")
```

```python
import joblib

name, model = best_model

joblib.dump(model, 'demand_model.pkl')
joblib.dump(pca, 'pca.pkl')
joblib.dump(scaler, 'scaler.pkl')

# Save log-transform flag so UI knows to use expm1
import json
with open('model_config.json', 'w') as f:
    json.dump({'model_name': name, 'log_transform': True}, f)

print(f"Saved best model: {name}")
```

## Rerunning for creating .pkl files

```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import numpy as np
import pandas as pd
import joblib
import json
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ── 1. Load matched data ──────────────────────────────────────────────────────
matched_df = pd.read_csv('matched_data.csv')
print(f"Loaded {len(matched_df)} rows")

# ── 2. Extract embeddings ─────────────────────────────────────────────────────
model_cnn = models.mobilenet_v2(pretrained=True)
model_cnn.classifier = torch.nn.Identity()
model_cnn.eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

def get_embedding(image_path):
    try:
        img = Image.open(image_path).convert('RGB')
        t = transform(img).unsqueeze(0)
        with torch.no_grad():
            return model_cnn(t).squeeze().numpy()
    except Exception as e:
        print(f"  Failed: {image_path} → {e}")
        return None

print("\nExtracting embeddings...")
embeddings = []
valid_indices = []

for idx, row in matched_df.iterrows():
    emb = get_embedding(row['image_path'])
    if emb is not None:
        embeddings.append(emb)
        valid_indices.append(idx)

matched_df = matched_df.loc[valid_indices].reset_index(drop=True)
embeddings = np.array(embeddings)
print(f"Embeddings shape: {embeddings.shape}")

# ── 3. Save embeddings + final df ─────────────────────────────────────────────
np.save('embeddings.npy', embeddings)
matched_df['embedding_index'] = range(len(matched_df))
matched_df.to_csv('matched_data_final.csv', index=False)
print("Saved: embeddings.npy, matched_data_final.csv")

# ── 4. Prepare features ───────────────────────────────────────────────────────
extra_features = matched_df[['avg_rate']].fillna(0).values
X = np.hstack([embeddings, extra_features])
y_log = np.log1p(matched_df['total_qty'].values)

pca = PCA(n_components=30)
X_reduced = pca.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_reduced)

# ── 5. Train GradientBoosting (best from earlier) ─────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_log, test_size=0.2, random_state=42
)

rf_model = RandomForestRegressor(
        n_estimators=200,
        max_depth=5,
        min_samples_leaf=4,
        max_features='sqrt',
        random_state=42
    )
rf_model.fit(X_train, y_train)

from sklearn.metrics import mean_absolute_error, r2_score
y_pred_log = rf_model.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_test_actual = np.expm1(y_test)

print(f"\nFinal Model Results:")
print(f"  MAE : {mean_absolute_error(y_test_actual, y_pred):.2f}")
print(f"  R²  : {r2_score(y_test_actual, y_pred):.4f}")

# ── 6. Save everything ────────────────────────────────────────────────────────
joblib.dump(rf_model, 'demand_model.pkl')
joblib.dump(pca,      'pca.pkl')
joblib.dump(scaler,   'scaler.pkl')

with open('model_config.json', 'w') as f:
    json.dump({'model_name': 'RandomForest', 'log_transform': True}, f)

print("\n✅ All files saved:")
print("   demand_model.pkl")
print("   pca.pkl")
print("   scaler.pkl")
print("   model_config.json")
print("   embeddings.npy")
print("   matched_data_final.csv")
```
